```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

+ 数据准备Series

```python
data=pd.Series(np.random.randn(1000),index=np.arange(1000))
data =data.cumsum()
data
```

```python
data.plot()
plt.show()
```

+ 数据准备DataFrame

```python
data = pd.DataFrame(np.random.randn(1000, 4), index=np.arange(1000), columns=list("ABCD"))
data = data.cumsum()
data
```

```python
data.head()  # 打印前五行数据
```

```python
data.plot()
plt.show()
```

+ plot()的参数
    * 'bar', 'hist', 'box', 'kde', 'area', scatter', hexbin', 'pie'等

### plot.scatter()打印数据点

```python
ax = data.plot.scatter(x='A', y='B', color='DarkBlue', label="Class 1")
data.plot.scatter(x='C', y='D', color='LightGreen', label='Class 2', ax=ax)
plt.show()
```

+ 把第一个ax 附在第二个plot中 这样就在一张图上打印了两组scatter
