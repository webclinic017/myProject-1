```python
import pandas as pd 
import numpy as np
```

```python
dates = pd.date_range('20130101', periods=6)
df1 = pd.DataFrame(np.arange(24).reshape((6,4)), index=dates, columns=['A', 'B', 'C', 'D'])
df1
```

# 直接改变表中某个值

```python
df1.iloc[2,2]=-1
df1.loc['20130101','A']=-1
df1
```

# 判断条件改变表中某个值 （用的比较多）

```python
df1.A[df1.A>0]=0  # 只把A这列中大于0的数变成0
df1
```

```python
df1[df1.B>1]=0  # 把符合B这列大于1的 所有行的数 都变成0
df1
```

# 添加一列

### 空列的添加

```python
df1['F']=0
df1['G']=True
df1['H']=None
df1['I']=np.nan
df1
```

### 已定义序列的添加

```python
df1['J']  = pd.Series([1,2,3,4,5,6], index=pd.date_range('20130101', periods=6))
df1
```
