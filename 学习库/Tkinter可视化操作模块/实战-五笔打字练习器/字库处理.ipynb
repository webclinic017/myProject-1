```python
f = open('xima.txt', 'r', encoding='utf-8')
dict01 = f.read()
f.close()

dict01 = dict01.split('\n')
len(dict01),type(dict01),dict01[0:100]
```

```python
xima_dict = {}
for i in range(len(dict01)):
    kong = dict01[i].find(' ')
    value = dict01[i][0:kong]
    k = dict01[i][kong+1:]

    key = k.split(' ')
    for j in range(len(key)):
        xima_dict[key[j]] = value
xima_dict,len(xima_dict)
```
