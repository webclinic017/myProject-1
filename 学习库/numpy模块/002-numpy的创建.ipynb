```python
import numpy as np
```

```python
a = np.array([2,23,4],dtype=np.int64)
```

dtype=np.int 默认是32位  
dtype=np.int64 是64位  一般来说我们用的是64


```python
print(a)
```

array 的打印输出与列表不同  
没有逗号

```python
print(a.dtype)
```

```python
# 多维矩阵
b = np.array([
    [2,23,4],
    [2,32,4]
])
print(b)
```

制作一个3行四列的矩阵 zeros矩阵 ones矩阵 empty矩阵

```python
c = np.zeros((3,4))
print(c)
```

```python
d = np.ones((3,4),dtype=np.int64)
print(d)
```

```python
e = np.empty((3,4),dtype=np.float64)
print(e)
```

制作一个有序的序列

```python
f = np.arange(10,20,2)
print(f)
```

制作一个有序的序列 矩阵化

```python
f = np.arange(10,34,2).reshape(3,4)
print(f)
```

先线段化 再矩阵化 开头是1结尾是10 总共12段 

```python
f = np.linspace(1,10,12,dtype=np.float64).reshape(3,4)
print(f)
```
