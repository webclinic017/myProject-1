```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('http://www.baidu.com/')
```

# 新建窗口

```python
newTab = 'window.open("https://www.sogou.com/");window.open("https://www.so.com/");window.open("https://cn.bing.com/");'
wd.execute_script(newTab)
```

# 句柄-handle

```python
for handle in wd.window_handles:
    print('窗口句柄：',handle)
```

# 切换窗口
+ 循环法
通常最终要的网页我们设置为mainWindow

```python
startWindow = wd.current_window_handle
# 虽然新打开了窗口，停留在bing上 但是， wd.current_window_handle并没有切换还是百度
for handle in wd.window_handles:
    wd.switch_to.window(handle)
    if '360' in wd.title:
        mainWindow = handle
        
        break
```

# 切换窗口
+ 顺序法


```python
wd.switch_to_window(wd.window_handles[-1])
print(wd.title)
```

```python
wd.switch_to_window(startWindow)
print(wd.title)
```

```python
wd.switch_to_window(mainWindow)
print(wd.title)
```

# 关闭标签页
### 关闭当前标签页

```python
wd.close()
# 关闭之后窗口句柄还是当前的窗口   需要自己去切换窗口
```

### 关闭所有标签页

```python
wd.quit()
```
