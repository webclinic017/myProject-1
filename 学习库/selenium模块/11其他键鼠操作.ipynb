除了常规的选择元素、点击元素、输入字符串 等操作。  
还有比如 鼠标右键点击、双击、移动鼠标到某个元素、鼠标拖拽等。  
这些可以通过ActionChains 类来实现。

# 例子：鼠标悬停某元素上

```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('https://www.qcc.com/')
```

### 导入ActionChains

```python
from selenium.webdriver.common.action_chains import ActionChains
ac = ActionChains(wd) # AChains类的实例化
```

```python
element1 = wd.find_element_by_css_selector('header:nth-child(3) [onmouseenter="getApplicationList();"]')
print(element1.text)
```


### 神奇代码：

```python
ac.move_to_element(element1).perform() # 蛮神奇的哦！！
```
