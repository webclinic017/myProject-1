## 设置最长等待时间
**wd.implicitly_wait(10)**
+ 这句代码表示如果执行不下去 就每半秒试一次 最长等待10秒

```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(10)
wd.get('https://www.baidu.com')

element = wd.find_element_by_id('kw')

element.send_keys('白月黑羽\n')

# id 为 1 的元素 就是第一个搜索结果
element = wd.find_element_by_id('1')

# 打印出 第一个搜索结果的文本字符串
print (element.text)
```

选择元素：  就是告诉浏览器，这个元素它有什么与众不同的地方，可以让浏览器一下子找到它。

鼠标选中元素 右键 审查元素

```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('https://www.qcc.com/')
```

### 根据id选择元素
根据规范， 如果元素有id属性 ，这个id 必须是当前html中唯一的。



```python
element = wd.find_element_by_id('searchkey')
```

```python
element.send_keys('浙江冠军食品有限公司\n')
```

### 根据 tag 名 选择元素


```python
elements = wd.find_elements_by_tag_name('li')
for element in elements:
    print(element.text)
```

### 根据 class属性 选择元素


```python
elements = wd.find_elements_by_class_name('item')
print(len(elements))
for element in elements:
    print(element.text)
```

```python
elements = wd.find_elements_by_class_name('pills-after')
print(len(elements))
for element in elements:
    print(element.text)
```

+ 先找到一个webelement，再从webelement中找到符合条件的元素

```python
A = wd.find_element_by_class_name('app-search')
print(A.text)
print('*********************************************************************************************************************************')
B = A.find_elements_by_class_name('pills-after')
print('B的长度',len(B))
for b in B:
    print(b.text)
```

```python
elements = wd.find_elements_by_class_name('frtrt')
if elements:
    target = elements[0].find_elements_by_tag_name('span')
    if target[1].text == '存续':
        print('True')
    else:
        print('False')
else:
    print('公司不存在')
```
