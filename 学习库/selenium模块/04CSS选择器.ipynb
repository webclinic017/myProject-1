css选择器就是一种语法 告诉浏览器 用什么样子来呈现 选定的元素  
它本就在html中就已经广泛应用 所以我们可以通过css来选择我们自己的元素

```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('https://www.qcc.com/')
element = wd.find_element_by_id('searchkey')
element.send_keys('一二三\n')
```

# class的css写法


```python
A   =     wd.find_elements_by_css_selector('.pills-after')
# 等价于  wd.find_elements_by_class_name('pills-after')
for a in A:
    print(a.text)
```

# tag的css写法


```python
A   =     wd.find_elements_by_css_selector('li')
# 等价于  wd.wd.find_elements_by_tag_name('li')
print(len(A))
for a in A:
    print(a.text)
```

# ID的css写法


```python
A   =     wd.find_element_by_css_selector('#searchKey')
# 等价于  wd.find_element_by_id('searchKey')
A.send_keys('啦啦啦')
```

## 孙元素的写法


```python
A = wd.find_elements_by_css_selector('.pills-after > .fitem > .fcheck > span')
for a in A:
    print('------------------------------------------------------------------------')
    print(a.text)
print('========================================================================')

B = wd.find_elements_by_css_selector('.pills-after      .fcheck       span')
for b in B:
    print('------------------------------------------------------------------------')
    print(b.text)
```

# 根据属性选择元素



```python
A = wd.find_element_by_css_selector('[href="https://www.qcc.com/cms?id=13"]')
print(A.text)
```

# 按次序选择子节点







```python
A = wd.find_elements_by_css_selector('.app-cascader-inline:nth-child(2) .pills-item')
for a in A:
    print('------------------------------------------------------------------------')
    print(a.text)
print('========================================================================')
```

# 兄弟节点选择



```python
A = wd.find_elements_by_css_selector('.about > h4+ul >li >a')
for a in A:
    print('------------------------------------------------------------------------')
    print(a.text)
print('========================================================================')
```

# 包含关系的筛选
## 一.css语法
### 1. A属性包含 'aaa'
div[A *= "aaa"]
### 2. A属性以 'aaa'开头
div[A ^= "aaa"]
### 3. A属性以 'aaa'结尾
div[A $= "aaa"]
## 二.xpath语法
### 1. B属性包含 'bbb'
//div[contains(@B,"bbb")]
### 2. B属性以 'bbb'开头
//div[starts-with(@B,"bbb")]
### 3. B属性以 'bbb'结尾 ------ 目前不支持
//div[starts-with(@B,"bbb")]
