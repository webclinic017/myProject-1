```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('http://cdn1.python3.vip/files/selenium/test2.html')
```

# select选择框

```python
a = wd.find_element_by_css_selector('#ss_single option[selected="selected"]')
print('默认选中的是：' + a.get_attribute('value'))
```

# select有专门的select类工具

```python
# 导入Select类
from selenium.webdriver.support.ui import Select

# 创建Select对象
s = Select(wd.find_element_by_css_selector('#ss_single'))
```

# 用 Select类 选中小雷老师
### 1.通过 可见文本 选择

```python
s.select_by_visible_text("小凯老师")
```

### 2.通过 value值  选择

```python
s.select_by_value("小雷老师")
```

### 3.通过 序号  选择 （0,1,2,3...）

```python
s.select_by_index(0)
```

# 用 Select类 操作多选框

```python
# 导入Select类
from selenium.webdriver.support.ui import Select

# 创建Select对象
ss = Select(wd.find_element_by_css_selector('#ss_multi'))
```

### 先清除所有 选中项

```python
ss.deselect_all()
```


### 1.通过 可见文本 选择

```python
ss.select_by_visible_text("小凯老师")
```

### 2.通过 value值  选择

```python
ss.select_by_value("小雷老师")
```

### 3.通过 序号  选择 （0,1,2,3...）

```python
ss.select_by_index(0)
```

# 取消选择

```python
ss.deselect_by_index(0)
ss.deselect_by_value("小雷老师")
ss.deselect_by_visible_text("小凯老师")
```
