```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('http://cdn1.python3.vip/files/selenium/test2.html')
```

# radio选择框
### 直接用WebElement的click方法，模拟用户点击就可以了。


```python
selected = wd.find_element_by_css_selector('#s_radio input[checked="checked"]')
print('默认选中的是：' + selected.get_attribute('value'))
```

### 改为 点击："小雷老师"

```python
select = wd.find_element_by_css_selector('#s_radio input[value="小雷老师"]')
select.click()
```
