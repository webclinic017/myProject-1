```python
from selenium import webdriver
wd = webdriver.Chrome('driver/chromedriver.exe')
wd.implicitly_wait(2)
wd.get('http://cdn1.python3.vip/files/selenium/test4.html')
```

### 操作一：

```python
operate1 = wd.find_element_by_id('b1')
operate1.click()
```


# alert类弹出框的处理



```python
# 打印弹出框的信息
t = wd.switch_to.alert.text
print(t)
```

```python
# 点击确定
wd.switch_to.alert.accept()
```

### 操作二:

```python
operate2 = wd.find_element_by_id('b2')
operate2.click()
```

```python
# 点击确定
wd.switch_to.alert.accept()
```

```python
operate2.click()
```

```python
# 点击取消
wd.switch_to.alert.dismiss()
```
